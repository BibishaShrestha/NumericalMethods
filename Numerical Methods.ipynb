## Solution of Non-linear Equations
#### For approximation of roots or zeros (solution) of non-linear equations

### 1. Bisection Method

```python
import math
def f(x):
    return 4*math.sin(x)-math.exp(x)
x1=int(input("Enter the 1st initial guess"))
x2=int(input("Enter the 2nd initial guess"))
e=float(input("Enter the error tolerence"))
n=int(input("Enter the no. of steps"))
while(f(x1)*f(x2)>=0):
    print("Guess again")
    x1=int(input("Enter the 1st initial guess"))
    x2=int(input("Enter the 2nd initial guess"))
print("\n Step \t\t x1 \t\t x2 \t\t x0 \t\t f(x0)")   
step=1
condition= True
while(condition):
    if (step>n):
        print("Not converged")
        break
    x0=(x1+x2)/2
    f0=f(x0)
    n=n+1
    print(step,"\t\t""%0.4f"%x1,"\t\t""%0.4f"%x2,"\t\t""%0.4f"%x0,"\t\t""%0.4f"%f(x0))
    if(math.fabs(f0)>e):
        condition=True
    else:
        condition=False
    if(f(x0)*f(x1)<0):
        x2=x0
    else:
        x1=x0
    step+=1
print("The value is:",x0)
    
```

### 2. Regular Falsi Method

```python
import math
def f(x):
    return 4*math.sin(x)-math.exp(x)
x1=int(input("Enter the 1st initial guess"))
x2=int(input("Enter the 2nd initial guess"))
e=float(input("Enter the error tolerence"))
n=int(input("Enter the no. of steps"))
while(f(x1)*f(x2)>=0):
    print("Guess again.")
    x1=int(input("Enter the 1st initial guess"))
    x2=int(input("Enter the 2nd initial guess"))
print("\n Step \t\t x1 \t\t x2 \t\t x0 \t\t f(x0)")   
step=1
condition= True
while(condition):
    if (step>n):
        print("Not converged")
        break
    if abs(f(x2)-f(x1))==1e-10:
        break
    x0=x1-(f(x1)*(x2-x1))/(f(x2)-f(x1))
    f0=f(x0)
    n=n+1
    print(step,"\t\t""%0.4f"%x1,"\t\t""%0.4f"%x2,"\t\t""%0.4f"%x0,"\t\t""%0.4f"%f(x0))
    if(math.fabs(f0)>e):
        condition=True
    else:
        condition=False
    if(f(x0)*f(x1)<0):
        x2=x0
    else:
        x1=x0
    step+=1
print("The value is:",x0)
    
```

### 4. Newton Raphson's Method

```python
import math
def f(x):
    return 4*math.sin(x)-math.e**x
def g(x):
    return 4*math.cos(x)-math.e**x
x0= float(input("Enter initial guess:"))
e= float(input("Enter the tolerance error:"))
n=int(input("Enter maximum no. of steps:"))
print("\n\n Newton Raphson's Method")
step=1
flag=1
condition=True
print("\nStep\t\t x0\t\t x1\t\t f(x1)")
while condition:
    if g(x0)== 0.0:
        print("Divide by zero error!")
        flag=0
        break
    x1= x0-f(x0)/g(x0)
    print(f"{step}\t\t {x0:.4f}\t\t {x1:.4f}\t\t {f(x1):.4f}")
    x0=x1
    step+=1
    if abs(f(x1))<=e:
        break
    if step>n:
        flag=0
        break
if flag ==1:
    print("\nThe root is: %.4f"%x1)
else:
    print("\n Not converged!")
```

## Interpolation
##### It is used to estimate unkown values of a function between a set of known data points.


#### A. For Equally Spaced Intervals:

### Forward Difference Table
##### For interpolation near the beginning of a data set

```python
import numpy as np

# Input the number of data points
n = int(input("Enter number of data points: "))

# Initialize arrays
x = np.zeros(n)
y = np.zeros((n, n))

# Input x and y values
print("Enter data for x and y:")
for i in range(n):
    x[i] = float(input(f"x[{i}] = "))
    y[i][0] = float(input(f"y[{i}] = "))

# Calculate forward difference table
for j in range(1, n):
    for i in range(n - j):
        y[i][j] = y[i + 1][j - 1] - y[i][j - 1]

# Display the forward difference table
print("\nFORWARD DIFFERENCE TABLE:\n")
for i in range(n):
    print(f"{x[i]:<8.2f}", end="")
    for j in range(n - i):
        print(f"{y[i][j]:<10.2f}", end="")
    print()
```

#### B. For Unequally Spaced Intervals:

### 1. Lagrange Interpolation Method

```python
import math
import numpy as np
n=int(input("Enter the no. of data:"))
x=np.zeros(n)
y=np.zeros(n)
x_val= float(input("Enter the value of x for which y is to be found:"))
print("Enter the value of x and y:")

for i in range(n):
    x[i]= float(input('x['+str(i)+']='))
    y[i]= float(input('y['+str(i)+']='))
result=0

for i in range(n):
    term=1
    for j in range(n):
        if i!=j:
            term*=((x_val-x[j])/(x[i]-x[j]))

    result= result+ term*y[i]
print("The interpolated value of %0.3f is %.3f"%(x_val,result))
```

### 2. Newton's Divided Difference Interpolation Method

```python
import numpy as np

n = int(input("Enter the number of data points: "))
x = np.zeros(n)
y = np.zeros((n, n))

# Input x and y
for i in range(n):
    x[i] = float(input(f"Enter x[{i}]: "))
    y[i][0] = float(input(f"Enter y[{i}]: "))

# Build divided difference table
for j in range(1, n):
    for i in range(n - j):
        y[i][j] = (y[i+1][j-1] - y[i][j-1]) / (x[i+j] - x[i])

x1 = float(input("Enter x to interpolate: "))

# Newton Interpolation
result = y[0][0]
term = 1

print("\nTerm\t| Coefficient\t| Term Value")
for i in range(1, n):
    term *= (x1 - x[i-1])
    coeff = y[0][i]
    term_val = coeff * term
    print(f"{i}\t| {coeff:.6f}\t| {term_val:.6f}")
    result += term_val

print(f"\nInterpolated value at x = {x1} is y = {result:.6f}")
```

## Numerical Integration
##### To find the definite Integral of a function

### 1. Trapezoidal Rule

```python
def f(x):
    return 1+ x**3
a=int(input("Enter the lower limit"))
b=int(input("Enter the upper limit"))
n=int(input("Enter the no. of subintervals"))
h=(b-a)/n
I=f(a)+f(b)
for i in range (1,n):
    I+=2*f(a+i*h)
I=I*h*0.5
print("I=",I)
```

### 2. Simpson's 1/3 Rule

```python
def f(x):
    return 1+x**3
a=int(input("Enter the lower limit"))
b=int(input("Enter the upper limit"))
n=int(input("Enter the no. of subintervals"))
if(n%2!=0):
    print("n to be even")
    exit()
h=(b-a)/n
I=f(a)+f(b)
for i in range (1,n):
    if(i%2==0):
        I+=2*f(a+i*h)
    else:
        I+=4*f(a+i*h)
I=(I*h)/3
print("I=",I)
```

### 3. Simpson's 3/8 Rule

```python
def f(x):
    return 1+x**3
a=int(input("Enter the lower limit"))
b=int(input("Enter the upper limit"))
n=int(input("Enter the no. of subintervals"))
if(n%3!=0):
    print("n to be multiple of 3")
    exit(1)
h=(b-a)/n
I=f(a)+f(b)
if(n%3==0):
    for i in range (1,n):
     if(i%3==0):
         I+=2*f(a+i*h)
     else:
         I+=3*f(a+i*h)
    I=(3*I*h)/8
    print("I=",I)
```

## Solution of Ordinary Differential Equations (ODE)
#### Initial Value Problems

### 1. Euler's Method (1st Order Method)

```python
def func(x,y):
    return 1+3*x*x

def euler_method(x0,y0,xn,h):
    n=round((xn-x0)/h)
    print("\n x\t\t y\t\t m\t\t y_next")
    for i in range(n):
        m=func(x0,y0)
        yn=y0+m*h
        print("%.4f\t\t %.4f\t\t %.4f\t\t %.4f"%(x0,y0,m,yn))
        y0=yn
        x0=x0+h
    print("\n At x=%4f,y=%4f(approx)"%(xn,y0))
print("Enter the initial guess:")
x0=float(input("x0= "))
y0=float(input("y0= "))
print("Enter the value of xn:")
xn=float(input("xn= "))
print("Enter the stepsize(h)")
h=float(input("h= "))
euler_method(x0,y0,xn,h)
```

### 2. RK-2 Method (Runge–Kutta 2nd order)

```python
def func(x,y):
    return 1+3*x*x

def Rk2_method(x0,y0,xn,h):
    n=round((xn-x0)/h)
    print("\n x\t\t y\t\t k1\t\t k2\t\t y_next")
    for i in range(n):
        k1=h*func(x0,y0)
        k2=h*func(x0+h/2,y0+k1/2)
        yn=y0+k2
        print("%.4f\t\t %.4f\t\t %.4f\t\t %.4f\t\t %.4f"%(x0,y0,k1,k2,yn))
        y0=yn
        x0=x0+h
    print("\n At x=%4f,y=%4f(approx)"%(xn,y0))
print("Enter the initial guess:")
x0=float(input("x0= "))
y0=float(input("y0= "))
print("Enter the value of xn:")
xn=float(input("xn= "))
print("Enter the stepsize(h)")
h=float(input("h= "))
Rk2_method(x0,y0,xn,h)
```

### 3. RK-4 Method(Runge–Kutta 4th order)

```python
def func(x,y):
    return 1+3*x*x

def Rk2_method(x0,y0,xn,h):
    n=round((xn-x0)/h)
    print("\n x\t\t y\t\t k1\t\t k2\t\t y_next")
    for i in range(n):
        k1=h*func(x0,y0)
        k2=h*func(x0+h/2,y0+k1/2)
        k3=h*func(x0+h/2,y0+k2/2)
        k4=h*func(x0+h,y0+k3)
        yn=y0+(k1+2*k2+2*k3+k4)/6
        print("%.4f\t\t %.4f\t\t %.4f\t\t %.4f\t\t %.4f\t\t %.4f\t\t %.4f"%(x0,y0,k1,k2,k3,k4,yn))
        y0=yn
        x0=x0+h
    print("\n At x=%4f,y=%4f(approx)"%(xn,y0))
print("Enter the initial guess:")
x0=float(input("x0= "))
y0=float(input("y0= "))
print("Enter the value of xn:")
xn=float(input("xn= "))
print("Enter the stepsize(h)")
h=float(input("h= "))
Rk2_method(x0,y0,xn,h)
```

### Least Square Method
##### It is used to find the best fitting line or curve through a set of data points by minimizing the sum of the squares of the vertical distances(residuals) between the observed and predicted values.

```python
import numpy as np

n = int(input("Enter the number of data:"))
x = np.zeros(n)
y = np.zeros(n)
sumx = 0
sumy = 0
sumxx = 0
sumxy = 0

for i in range(n):
    x[i] = float(input(f"x[{i}] = "))
    y[i] = float(input(f"y[{i}] = "))
    sumx += x[i]
    sumy += y[i]
    sumxx += x[i] * x[i]
    sumxy += x[i] * y[i]

# Compute slope (b) and intercept (a) after the loop
b = (n*sumxy - sumx*sumy) / (n*sumxx - sumx*sumx)
a = (sumy - b*sumx) / n

print(f"\nThe equation of best fit line is: y = {a:.4f} + {b:.4f}x")
```

### Power Method
##### It is iterative method used to find the largest eigenvalue and eigenvector of a matrix.

```python
import numpy as np

# Input matrix
n = int(input("Enter the order of the square matrix: "))
A = np.zeros((n, n))
print("Enter the matrix elements row-wise:")
for i in range(n):
    for j in range(n):
        A[i][j] = float(input(f"A[{i}][{j}] = "))

# Initial guess vector
x = np.ones(n)
e = float(input("Enter the tolerance error: "))
max_iter = int(input("Enter maximum number of iterations: "))

print("\nStep\tEigenvalue (lambda)\t\tEigenvector (x)")

for step in range(1, max_iter + 1):
    # Multiply matrix by vector
    x_new = np.dot(A, x)

    # Estimate dominant eigenvalue (use any non-zero element ratio)
    lambda_ = max(abs(x_new))
    
    # Normalize the vector
    x_new = x_new / lambda_

    print(f"{step}\t{lambda_:.6f}\t\t\t{x_new}")

    # Check convergence
    if np.allclose(x, x_new, atol=e):
        break

    x = x_new

print(f"\nDominant eigenvalue: {lambda_:.6f}")
print(f"Corresponding eigenvector: {x_new}")
```

## Solution of System of Linear Algebric Equations

### Gauss Elimination Method (Direct Method)

```python
import numpy as np

# Input number of equations
n = int(input("Enter the number of equations: "))

# Augmented matrix A|b
a = np.zeros((n, n+1))
print("Enter the augmented matrix coefficients row-wise:")
for i in range(n):
    for j in range(n+1):
        a[i][j] = float(input(f"a[{i}][{j}] = "))

# Forward Elimination
for i in range(n):
    # Pivoting if diagonal element is zero
    if a[i][i] == 0.0:
        for k in range(i+1, n):
            if a[k][i] != 0:
                a[[i, k]] = a[[k, i]]  # Swap rows
                break
        else:
            print("Matrix is singular or inconsistent.")
            exit()

    for j in range(i+1, n):
        ratio = a[j][i] / a[i][i]
        a[j] = a[j] - ratio * a[i]

# Back Substitution
x = np.zeros(n)
for i in range(n-1, -1, -1):
    x[i] = (a[i][n] - np.dot(a[i][i+1:n], x[i+1:n])) / a[i][i]

# Output solution
print("\nThe solution is:")
for i in range(n):
    print(f"x[{i}] = {x[i]:.6f}")
```
